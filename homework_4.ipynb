```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Players.csv', usecols = [1,2,3,4,5,6,7])
df.head()
```

```python
df1 = pd.read_csv('player_data.csv')
df1.head()
```

```python
df1 = df1.drop(columns=['weight','height','birth_date'])
df1.head()
```

```python
df = pd.merge(df, df1, how = 'inner', left_on= ['Player', 'collage'], right_on=['name', 'college'])
df.head()
```

```python
df=df.drop(columns=['name'])
df.head()
```

```python
df['weight'].plot.box()
```

На цьому графіку бачимо досить велику кількість викидів по вазі гравців.

```python
df['height'].plot.box()
```

На цьому графіку бачимось відносно невелику кількість викидів по зрості гравців.

```python
sns.kdeplot(df['born'])
```

На даному графіку бачимо розподіл по народжуваності гравців по рокам, також бачимо спад перед другою світовою та у 80-ті роки.

```python
df.plot.hexbin(x='height', y='weight', gridsize=25, cmap='Blues')
plt.title('Взаємозв’язок зросту і ваги гравців')
```

Тут бачимо загальний взаємозвязок між зростом та вагою гравців, та приблизні значення, який є найбільше в нашому датасеті

```python
sns.kdeplot(df.loc[:,['height','weight']],fill=True)
```

Тут бачимо розподіл значень ваги та росту гравців, що показує не зовсім номральний розподіл.

```python
sns.kdeplot(df.loc[:,['year_start','year_end']],fill=False)
```

Бачимо сильну залежність між роками починання гри у гравців та закінченням карєри, сильний спад з кінця 1955-х років до 1965-ого року, а також перед 2020 роком найбільший спад, ймовірно через коронавірус

```python
sns.countplot(df['position'])
```

Бачимо розподів гравців за позиціями, і те що більшість згравці грали на позиціях F або G

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, diagonal="kde");
```

Бачимо загальний розподіл всіх значень між одна одними, а також бачимо сильну кореляцію в декількох ознаках

```python
corr_matrix = df.corr(numeric_only=True)
import seaborn as sns
sns.heatmap(corr_matrix)
```

Тут бачимо що більшість ознак сильно корелює між собою, це рік народження, старт карєри та кінець карєри. А також погранична кореляція між вагою та зростом.

```python
sns.relplot(data=df, x="weight", y="height", hue="position")
```

Бачимо наскільки залежить зріст та вага гравця від його позиції

```python
df['years_of_experience'] = df['year_end'] - df['year_start']
sns.displot(df['years_of_experience'])
```

Створили фічу скільки гравець відіграв років

```python
df['years_started'] = df['year_start'] - df['born']
sns.displot(df['years_started'])
```

```python
df['years_started'].plot.box()
```

Багато викидів в нової фічі, яка означає в скільки років гравець почав карєру

```python
df['middle_of_career'] = (df['year_end']+df['year_start'])/2
sns.displot(df['middle_of_career'])
```

Нова фіча На який вік Припала середина карєри

```python
df['year_in_end'] = df['year_end'] - df['born']
sns.displot(df['year_in_end'])
```

Нова фіча Скільки було років коли гравець закінчив карєру

```python
df['weight_and_height']= df['weight']+df['height']
```

```python
sns.displot(df['weight_and_height'])
```

Нова фіча, сума ваги гравця та його зросту
