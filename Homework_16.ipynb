```python
from transformers import MarianMTModel, MarianTokenizer
from transformers import pipeline

pipeline = 'translate'
model_name = "Helsinki-NLP/opus-mt-en-uk"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)


text = """Spain’s PM says U.S. position is ‘untrue,’ EU doesn’t apply 39% tariffs

Spain’s Prime Minister Pedro Sanchez slammed the U.S. for the 20% “reciprocal” tariff duty it imposed on EU imports to the States.

The U.S. said it was imposing a 20% duty on EU imports to the States, claiming that the measure was a “reciprocal” tariff as the bloc had a 39% import tariff on U.S. goods.

Prime Minister Sanchez said the 39% figure was “untrue” and added that a return to “19th century protectionism is not an intelligent move,” in comments reported by Reuters Thursday."""


sentences = text.split('\n')


translated_parts = []
for sentence in sentences:
    if sentence.strip():
        inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=512)
        translated = model.generate(**inputs)
        translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
        translated_parts.append(translated_text)


translated_text = '\n\n'.join(translated_parts)
print("Переклад:")
print(translated_text)
```

```python
from transformers import pipeline

def translate_text(text, src_lang="eng_Latn", tgt_lang="ukr_Cyrl", model_name="facebook/nllb-200-1.3B", max_length=200):

    translator = pipeline(
        "translation",
        model=model_name,
        src_lang=src_lang,
        tgt_lang=tgt_lang,
    )

    result = translator(text, max_length=max_length)
    return result[0]['translation_text']

if __name__ == "__main__":
    input_text = """
    The U.S. Department of Energy announced new funding for AI research.
    Scientists believe this will accelerate breakthroughs in medicine.
    """

    translated_text = translate_text(input_text)
    print("Оригінальний текст:")
    print(input_text)
    print("\nПереклад:")
    print(translated_text)
```

```python
import torch
translator = pipeline(
    "translation",
    model="facebook/nllb-200-1.3B",
    src_lang="eng_Latn",
    tgt_lang="ukr_Cyrl",
    device="cuda" if torch.cuda.is_available() else "cpu"
)

texts = [
    "The government announced new reforms.",
    "Climate change is a global challenge.",
    "Ukrainian culture is rich and diverse."
]

# закидуєм batch
translated = translator(texts, max_length=100, batch_size=3)
for original, translation in zip(texts, translated):
    print(f"{original} = {translation['translation_text']}")
```
